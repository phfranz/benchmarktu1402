```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shapes = np.transpose(np.loadtxt("Job-1_modes.dat"))
shapes.shape
```

```python
x_values = np.zeros([5,20])
y_values = np.zeros([5,20])
```

```python
column = 0
for i in range(0,120,6):

    x_values[:,column] = shapes[:,i]
    y_values[:,column] = shapes[:,i+1]

    column += 1
```

```python
coordinates = np.transpose(np.loadtxt("Output_nodes.dat"))

x_coordinates=[]

for i in range(0,60,3):
        
        x_coordinates.append(coordinates[1,i])
```

```python
freqz = np.loadtxt("Job-1_frequencies.dat")
```

```python
# mode shape 1

fig, ax = plt.subplots(figsize=[12,8])

fig.suptitle('Mode shapes computed with Python')

plt.plot(x_coordinates, y_values[0,:],  label = str(np.round(freqz[0],3))+' Hz')
plt.plot(x_coordinates, y_values[1,:],  label = str(np.round(freqz[1],3))+' Hz')
plt.plot(x_coordinates, y_values[2,:],  label = str(np.round(freqz[2],3))+' Hz')
plt.plot(x_coordinates, y_values[3,:],  label = str(np.round(freqz[3],3))+' Hz')
plt.plot(x_coordinates, y_values[4,:],  label = str(np.round(freqz[4],3))+' Hz')
plt.legend()

ax.set_xlabel('Length of two-span beam [m]')
ax.set_ylabel('Mode Shapes')
```

```python
fig, ax = plt.subplots(figsize=[12,8])

fig.suptitle('Mode shapes computed with Python')

j = 2
plt.plot(x_coordinates, y_values[j,:],  label = str(np.round(freqz[j],3))+' Hz')
plt.legend()

ax.set_xlabel('Length of two-span beam [m]')
ax.set_ylabel('Mode Shapes')
```
